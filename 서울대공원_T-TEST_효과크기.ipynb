```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
```

```python
data_2018 = pd.read_csv('2018_서울대공원_입장객(강수여부추가ver.).csv', encoding='utf-8-sig')
```

```python
rainy_visitors_2018 = data_2018[data_2018['is_rainy'] == '비옴']['visitors'].dropna()
non_rainy_visitors_2018 = data_2018[data_2018['is_rainy'] == '비안옴']['visitors'].dropna()

mean_rainy_2018 = rainy_visitors_2018.mean()
mean_non_rainy_2018 = non_rainy_visitors_2018.mean()
std_rainy_2018 = rainy_visitors_2018.std()
std_non_rainy_2018 = non_rainy_visitors_2018.std()

pooled_std_2018 = np.sqrt(((std_rainy_2018**2) + (std_non_rainy_2018**2)) / 2)
cohens_d_2018 = (mean_non_rainy_2018 - mean_rainy_2018) / pooled_std_2018

{
    "비오는 날 평균 방문객 수": mean_rainy_2018,
    "비오지 않는 날 평균 방문객 수": mean_non_rainy_2018,
    "Cohen's d (효과 크기)": cohens_d_2018
}
```

```python
rainy_visitors_2018 = data_2018[data_2018['is_rainy'] == '비옴']['visitors'].dropna()
non_rainy_visitors_2018 = data_2018[data_2018['is_rainy'] == '비안옴']['visitors'].dropna()

t_stat_2018, p_value_2018 = ttest_ind(rainy_visitors_2018, non_rainy_visitors_2018, equal_var=False)

{
    "비오는 날 평균 방문객 수": rainy_visitors_2018.mean(),
    "비오지 않는 날 평균 방문객 수": non_rainy_visitors_2018.mean(),
    "t-statistic": t_stat_2018,
    "p-value": p_value_2018,
}
```
