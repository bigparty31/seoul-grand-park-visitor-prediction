```python
import pandas as pd
```

```python
visitors_2023 = pd.read_csv('서울대공원_입장객_정보_2023년_day_한글복원.csv', encoding='euc-kr')
```

```python
data = visitors_2023 
data
```

```python
from datetime import datetime

# Define the basic Korean public holidays for 2023 (including substitute holidays)
korean_holidays = [
    '2023-01-01',  # New Year's Day
    '2023-01-21', '2023-01-22', '2023-01-23',  # Lunar New Year and substitute holiday
    '2023-03-01',  # Independence Movement Day
    '2023-05-05',  # Children's Day
    '2023-05-27',  # Substitute holiday for Buddha's Birthday
    '2023-06-06',  # Memorial Day
    '2023-08-15',  # Liberation Day
    '2023-09-28', '2023-09-29', '2023-09-30', '2023-10-02',  # Chuseok and substitute holiday
    '2023-10-03',  # National Foundation Day
    '2023-10-09',  # Hangul Day
    '2023-12-25',  # Christmas Day
]

# Convert to datetime for comparison
korean_holidays = [datetime.strptime(date, '%Y-%m-%d').date() for date in korean_holidays]

# Add a new column classifying into the 4 categories
def classify_day(row):
    date = datetime.strptime(row['date'], '%Y-%m-%d').date()
    if date in korean_holidays:
        return '공휴일'
    elif row['day'] == '토':
        return '주말'
    elif row['day'] == '일':
        return '주말'
    else:
        return '평일'

data['category'] = data.apply(classify_day, axis=1)

data
```

```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd

# Step 1: 학습 데이터와 테스트 데이터 준비
# 학습 데이터: 2019년부터 2022년까지
train_data = all_data_filtered[(all_data_filtered['date'] >= '2019-01-01') & (all_data_filtered['date'] <= '2022-12-31')]

# 테스트 데이터: 2023년
test_data = all_data_filtered[(all_data_filtered['date'] >= '2023-01-01') & (all_data_filtered['date'] <= '2023-12-31')]

# Prophet 모델이 요구하는 형식으로 데이터 변환 (date -> ds, visitors -> y)
train_prophet_data = train_data[['date', 'individual_visitors']].rename(columns={'date': 'ds', 'individual_visitors': 'y'})
test_prophet_data = test_data[['date', 'individual_visitors']].rename(columns={'date': 'ds', 'individual_visitors': 'y'})

# Step 2: Prophet 모델 초기화 및 학습
model = Prophet()  # Prophet 모델 생성
model.fit(train_prophet_data)  # 학습 데이터로 모델 학습

# Step 3: 미래 예측 데이터프레임 생성 (2023년 예측)
future = model.make_future_dataframe(periods=365)  # 미래 365일의 데이터를 예측
forecast = model.predict(future)  # 예측 수행

# Step 4: 2023년에 해당하는 예측값 추출
forecast_2023 = forecast[(forecast['ds'] >= '2023-01-01') & (forecast['ds'] <= '2023-12-31')]

# Step 5: 성능 평가 (MAE, RMSE 계산)
mae = mean_absolute_error(test_prophet_data['y'], forecast_2023['yhat'])  # 평균 절대 오차
rmse = np.sqrt(mean_squared_error(test_prophet_data['y'], forecast_2023['yhat']))  # 평균 제곱근 오차

# MAE와 RMSE 출력
print(f"평균 절대 오차 (MAE): {mae}")
print(f"평균 제곱근 오차 (RMSE): {rmse}")

# Step 6: 실제값 vs 예측값 시각화
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))

# 실제값 (2023년)
plt.plot(test_prophet_data['ds'], test_prophet_data['y'], label='실제값', color='blue')

# 예측값 (2023년)
plt.plot(forecast_2023['ds'], forecast_2023['yhat'], label='예측값', color='red')

# 예측 구간 (불확실성 구간) 시각화
plt.fill_between(
    forecast_2023['ds'],
    forecast_2023['yhat_lower'],
    forecast_2023['yhat_upper'],
    color='pink',
    alpha=0.3,
    label='불확실성 구간'
)

# 그래프 제목 및 축 설정
plt.title('2023년 방문객 수 실제값 vs 예측값')
plt.xlabel('날짜')
plt.ylabel('방문객 수')
plt.legend()  # 범례 표시
plt.grid()  # 격자 표시
plt.show()
```

```python
# 성능 평가 - MAPE 계산
def calculate_mape(y_true, y_pred):
    """
    MAPE(Mean Absolute Percentage Error)를 계산합니다.
    y_true: 실제값
    y_pred: 예측값
    """
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# MAPE 계산
mape_with_holidays = calculate_mape(test_prophet_data['y'], forecast_2023_with_holidays['yhat'])

# 예측 정확도 계산
accuracy = 100 - mape_with_holidays

# 결과 출력
print(f"평균 절대 퍼센트 오차 (MAPE): {mape_with_holidays:.2f}%")
print(f"예측 정확도: {accuracy:.2f}%")
```

```python
import numpy as np
import pandas as pd
from prophet import Prophet
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

# Step 1: 데이터 준비
# 평일 데이터 필터링
weekday_data = all_data_filtered[(all_data_filtered['category'] == '평일')]

# 학습 데이터 (2019~2022년)
train_weekday_data = weekday_data[(weekday_data['date'] >= '2019-01-01') & (weekday_data['date'] <= '2022-12-31')]

# 테스트 데이터 (2023년)
test_weekday_data = weekday_data[(weekday_data['date'] >= '2023-01-01') & (weekday_data['date'] <= '2023-12-31')]

# Prophet 데이터 형식에 맞게 변환 (date -> ds, individual_visitors -> y)
train_prophet_data = train_weekday_data[['date', 'individual_visitors']].rename(columns={'date': 'ds', 'individual_visitors': 'y'})
test_prophet_data = test_weekday_data[['date', 'individual_visitors']].rename(columns={'date': 'ds', 'individual_visitors': 'y'})

# Step 2: Prophet 모델 학습 및 예측
prophet_model = Prophet()
prophet_model.fit(train_prophet_data)

# Prophet 예측값 생성 (테스트 데이터와 날짜 매칭)
future = pd.DataFrame(test_weekday_data['date']).rename(columns={'date': 'ds'})
prophet_forecast = prophet_model.predict(future)

# Prophet 예측값 필터링
prophet_predictions = prophet_forecast[['ds', 'yhat']]

# 디버깅: Prophet 예측값과 테스트 데이터 크기 확인
print(f"Prophet 예측값 크기: {len(prophet_predictions)}")
print(f"테스트 데이터 크기: {len(test_weekday_data)}")

# 날짜 매칭 검토
if len(prophet_predictions) != len(test_weekday_data):
    raise ValueError("Prophet 예측값과 테스트 데이터의 크기가 일치하지 않습니다.")

# Step 3: 잔차 계산
prophet_residuals = test_weekday_data['individual_visitors'].values - prophet_predictions['yhat'].values

# 디버깅: 잔차 데이터 크기 확인
print(f"전체 잔차 데이터 크기: {len(prophet_residuals)}")
print(f"테스트 데이터 크기: {len(test_weekday_data)}")

# 테스트 데이터 전체를 Prophet 잔차로 사용
test_residuals = prophet_residuals.reshape(-1, 1)

# MinMaxScaler 적용
scaler = MinMaxScaler(feature_range=(0, 1))
test_residuals_scaled = scaler.fit_transform(test_residuals)

# 데이터 크기 확인
print(f"테스트 잔차 데이터 크기 (스케일링 후): {len(test_residuals_scaled)}")

# Step 4: 시퀀스 데이터 생성 함수
def create_sequences(data, sequence_length=30):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i + sequence_length])
        y.append(data[i + sequence_length])
    return np.array(X), np.array(y)

# 시퀀스 생성
sequence_length = 30
X_test, y_test = create_sequences(test_residuals_scaled, sequence_length)

# LSTM 입력 형태로 데이터 변환 (samples, time steps, features)
X_test = np.expand_dims(X_test, axis=-1)

# Step 5: LSTM 모델 구성
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(50, return_sequences=True, input_shape=(sequence_length, 1)),
    tf.keras.layers.LSTM(50),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

# 모델 학습
history = model.fit(X_test, y_test, epochs=20, batch_size=32, verbose=1)

# 잔차 예측
lstm_predictions_scaled = model.predict(X_test)

# 정규화 복원
lstm_predictions = scaler.inverse_transform(lstm_predictions_scaled)

# Step 6: 최종 예측 (Prophet + LSTM)
final_predictions = prophet_predictions['yhat'].values[sequence_length:] + lstm_predictions.flatten()

# 최종 성능 평가
y_true = test_weekday_data['individual_visitors'].values[sequence_length:]
mae = mean_absolute_error(y_true, final_predictions)
rmse = np.sqrt(mean_squared_error(y_true, final_predictions))

print(f"앙상블 모델 MAE: {mae}")
print(f"앙상블 모델 RMSE: {rmse}")

# Step 7: 시각화
plt.figure(figsize=(15, 6))
plt.plot(test_weekday_data['date'].iloc[sequence_length:].values, y_true, label='실제값', color='blue')
plt.plot(test_weekday_data['date'].iloc[sequence_length:].values, final_predictions, label='예측값 (앙상블)', color='red')
plt.title('앙상블 모델을 사용한 평일 방문객 수 예측 (Prophet + LSTM)')
plt.xlabel('날짜')
plt.ylabel('방문객 수')
plt.legend()
plt.grid()
plt.show()
```
